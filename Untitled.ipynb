```python
from preprocessing.normal_tree import NormalTree
from nltk.tree import Tree
with open('files/test.txt', 'r') as f:
    test = [NormalTree.normal_fromstring(x, lower=False) for x in f]
```

```python
from nltk.tag.stanford import StanfordPOSTagger
tagger = StanfordPOSTagger('./parsing/english-bidirectional-distsim.tagger', path_to_jar='parsing/stanford-postagger.jar')
```

```python
from nltk import pos_tag
```

```python
from tqdm import tqdm
```

```python
tot, right = 0, 0
for tree in tqdm(test[:100]):
    tagged = tagger.tag(tree.leaves())
    tot += len(tree.leaves())
    for i, x in enumerate(tree.pos()):
        if x[1][x[1].rfind('+')+1:] == tagged[i][1]:
            right += 1
```

```python
tot, right = 0, 0
for tree in tqdm(test):
    tagged = pos_tag(tree.leaves())
    tot += len(tree.leaves())
    for i, x in enumerate(tree.pos()):
        if x[1][x[1].rfind('+')+1:] == tagged[i][1]:
            right += 1
```
